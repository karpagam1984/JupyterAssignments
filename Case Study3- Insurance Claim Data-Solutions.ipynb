## 2. Perform a data audit for the datatypes and find out if there are any mismatch within the current datatypes of the columns and their business significance.

```python
insure_df
```

```python
# Assigning object datatype to categorical columns

for i in ['claim_id', 'customer_id', 'CUST_ID']:
    insure_df[i]= insure_df[i].astype(str)
```

```python
# Assigning date datatype to date columns

insure_df['claim_date']= pd.to_datetime(insure_df['claim_date'],format="%m/%d/%Y")
insure_df['DateOfBirth']=pd.to_datetime(insure_df['DateOfBirth'],format='%d-%b-%y')
```

## 3. Convert the column claim_amount to numeric. Use the appropriate modules/attributes to remove the $ sign.

```python
# Remove $ from the claim_amount column.

insure_df.claim_amount= insure_df.claim_amount.str.replace('$', '')
```

```python
# Assigning datatype to the claim_amount column

insure_df.claim_amount= insure_df.claim_amount.astype('float64')
```

```python
insure_df.claim_amount.info()
```

## 4. Of all the injury claims, some of them have gone unreported with the police. Create an alert flag (1,0) for all such claims.

```python
insure_df.loc[:, 'flag']= np.where(insure_df.police_report == 'No', 0, np.where(insure_df.police_report== 'Yes', 1, np.nan))
```

## 6. Check for missing values and impute the missing values with an appropriate value. (mean for continuous and mode for categorical)

```python
insure_df.isna().sum()
```

```python
for i in ['claim_amount', 'total_policy_claims', 'flag']:
    if insure_df[i].dtype== 'object':
        insure_df.fillna(insure_df[i].mode()[0], inplace= True)
    else:
        insure_df.fillna(insure_df[i].mean(), inplace= True)
```

```python
insure_df.isna().sum()
```

## 7. Calculate the age of customers in years. Based on the age, categorize the customers according to the below criteria

## Children < 18

## Youth 18-30

## Adult 30-60

## Senior > 60

```python
# Creating a new column 'Age'

insure_df['Age']= date.today().year - insure_df.DateOfBirth.dt.year
```

```python
insure_df.loc[:, 'criteria'] = np.where(insure_df.Age <18, 'Children', 
                                      np.where((insure_df.Age >= 18) & (insure_df.Age<= 30), 'Youth',
                                      np.where((insure_df.Age > 30) & (insure_df.Age <= 60), 'Adult', 'Senior')))
```

## 8. What is the average amount claimed by the customers from various segments?

```python
insure_df.pivot_table(values= 'claim_amount', index= 'Segment', aggfunc= 'mean')
```

## 9. What is the total claim amount based on incident cause for all the claims that have been done at least 20 days prior to 1st of October, 2018.

```python
insure_df[insure_df.claim_date<=(dt.datetime(2018, 10, 1)-dt.timedelta(days=20))].groupby('incident_cause')['claim_amount'].sum().round(2).add_prefix('Total_Claim_Amt_')
```
